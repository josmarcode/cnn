#### Importar dataset

```python
from datasets import load_dataset
from PIL import Image

# Load CIFAR-10 dataset
dataset = load_dataset("cifar10")

# Dictionary of class names
class_names = {
    0: "airplane",
    1: "automobile",
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck",
}
```

```python
# Visualize the first 10 images
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 10, figsize=(10, 1))
for i in range(10):
    ax[i].set_axis_off()
    ax[i].imshow(dataset["train"][i]["img"])
    # Add the label
    ax[i].set_title(class_names[dataset["train"][i]["label"]])
plt.show()
```

```python
# Get resume of dataset with pandas
import pandas as pd

# Create a dataframe with the train dataset
df = pd.DataFrame(dataset["train"])

# Add a column with the class name
df["class_name"] = df["label"].apply(lambda x: class_names[x])

# Get the number of images per class
df.groupby("class_name").count()
```
